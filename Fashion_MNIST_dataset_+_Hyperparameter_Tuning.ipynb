```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
fashion_mnist = keras.datasets.fashion_mnist
```

```python
(train_images,train_labels),(test_images, test_labels) = fashion_mnist.load_data()
```

```python
train_images = train_images/255.0
test_images = test_images/255.0
```

```python
train_images[0].shape
```

```python
train_images[0][:2]
```

```python
#first dimension is set to the length of the original array and 28*28; 1 repreents grayscale
train_images = train_images.reshape(len(train_images),28,28,1)
test_images = test_images.reshape(len(test_images),28,28,1)
```

```python
import tensorflow as tf
from keras import layers, models
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier


# Define the model
def create_model(filters1, kernel_size1, filters2, kernel_size2, dropout_rate=0.2):

    model = models.Sequential()
    model.add(layers.Conv2D(filters1, kernel_size1, activation='relu', input_shape=(28, 28, 1)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(filters2, kernel_size2, activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(10, activation='softmax'))  # Assuming 10 classes for classification

    # Compile the model
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# # Create KerasClassifier for use in GridSearchCV
model = KerasClassifier(model=create_model, epochs=3, batch_size=32, verbose=1)

# # Define hyperparameters to tune
# Define hyperparameters to tune
param_grid = {
    'model__filters1': [32, 64],
    'model__kernel_size1': [(3, 3), (5, 5)],
    'model__filters2': [64, 128],
    'model__kernel_size2': [(3, 3), (5, 5)],
    'model__dropout_rate': [0.2, 0.5]
}


# # Use GridSearchCV for hyperparameter tuning
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_result = grid.fit(train_images, train_labels)

# # Print the best parameters and corresponding accuracy
# print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# # Print the best parameters and corresponding accuracy
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
